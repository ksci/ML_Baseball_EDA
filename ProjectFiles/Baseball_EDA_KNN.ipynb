```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import StratifiedKFold
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, auc
from numpy import interp
df = pd.read_csv("Data\\2019Custom\\2019_JUNE_SEA_GameStats.csv", header=None)
df
```

```python
X = df.loc[1:,8:].values
y = df.loc[1:,3:3].values.ravel()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
scaler = StandardScaler()
X_transform = scaler.fit_transform(X)
le = LabelEncoder()
y_label = le.fit_transform(y)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
    train_test_split(X_transform, y_label, 
                     test_size=0.30,
                     stratify=y_label,
                     random_state=1)
```

```python
def print_pretty(predict, actual):
    from prettytable import PrettyTable
    t = PrettyTable(['Predicted', 'Actual', 'Truth'])
    for pred,act in zip(predict,actual):
        truth = (pred==act)
        t.add_row([pred, act,truth])
    print(t)
    
def get_accuracy(predict,actual):
    total = len(predict)
    truth = 0.
    for pred,act in zip(predict,actual):
        truth += (pred==act)
    return truth / total
```

```python
import sklearn.neighbors as sk
sk_NN = sk.KNeighborsClassifier(n_neighbors=6,weights='distance',algorithm='brute')
sk_NN.fit(X_train,y_train)
predict = sk_NN.predict(X_test)
print_pretty(predict,y_test)
print(f'accuracy:{round(get_accuracy(predict, y_test),4)}')
```

```python
for i in range(1, len(X_train)):
    sk_NN = sk.KNeighborsClassifier(n_neighbors=i,weights='distance',algorithm='kd_tree')
    sk_NN.fit(X_train,y_train)
    predict = sk_NN.predict(X_test)
    print_pretty(predict,y_test)
    print(f'accuracy for {i} NN:{round(get_accuracy(predict, y_test),4)}')
```
